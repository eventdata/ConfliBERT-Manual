```python
def extract_urls_topic(url):
    try:
      coverpage = requests.get(url).content
      soup = BeautifulSoup(coverpage, 'html.parser')
     
      rows = []

      for link in soup.find_all('article', class_='row'):   
        rows.append(link.a['href'])

    except:
      rows = 'error'
      
    return rows
```

```python
extract_urls_topic('https://www.elmostrador.cl/noticias/mundo/')
```

```python
pandarallel.initialize(nb_workers= 30, progress_bar=True)
df = pd.DataFrame(range(1,1000),columns = ['page'])
url_str=''

#mundo
for i in range(0,999) :
    url_str='https://www.elmostrador.cl/noticias/mundo/page/'+str(i)
    df.page[i] = url_str


print(len(df))
df['rows']=df['page'].parallel_apply(extract_urls_topic)
df = df[df.rows.str.len()!=0].reset_index(drop=True)
df = df.explode('rows', ignore_index=True)
df=df.drop_duplicates(subset='rows')
df = df.dropna()
print(df)
df.to_csv('/content/drive/MyDrive/Colab Notebooks/elmonstrado/url.csv',index=False)
```
