```python
pandarallel.initialize(nb_workers= 30, progress_bar=True)
df = pd.DataFrame({'https://www.diariocordoba.com/sitemap_index_9665f.xml'},columns = ['page'])
try:
  coverpage = requests.get('https://www.diariocordoba.com/sitemap_index_9665f.xml').content
  soup = BeautifulSoup(coverpage, 'html.parser')
  rows = []
  coverpage=soup.find_all('loc')
  for link in coverpage:   
    rows.append(link.get_text())
except:
  rows = 'error'
df['rows']=[rows]
df = df.explode('rows', ignore_index=True)
df=df.drop_duplicates(subset='rows')
df = df.dropna()
print(df)
```

```python
def extract_urls_topic(url):
    try:
      coverpage = requests.get(url).content
      soup = BeautifulSoup(coverpage, 'html.parser')
     
      rows = []
      coverpage=soup.find_all('loc')
      for link in coverpage:  
        if(link.get_text().find('internacional')>=0):
          rows.append(link.get_text())
        else:
           if(link.get_text().find('espana')>=0):
            rows.append(link.get_text())
           else:
              if(link.get_text().find('andalucia')>=0):
                rows.append(link.get_text())
    except:
      rows = 'error'
      
    return rows
```

```python
extract_urls_topic('https://www.diariocordoba.com/sitemap_contents_2017_11_9665f_001.xml')
```

```python
pandarallel.initialize(nb_workers= 30, progress_bar=True)

df['rows_article']=df['rows'].parallel_apply(extract_urls_topic)
df = df.explode('rows_article', ignore_index=True)
df=df.drop_duplicates(subset='rows_article')
df = df.dropna()
print(df)
df.to_csv('/content/drive/MyDrive/Colab Notebooks/diariocordoba/url.csv',index=False)
```
