```python
def extract_urls_topic(url):
    try:
      coverpage = requests.get(url).content
      soup = BeautifulSoup(coverpage, 'html.parser')
     
      rows = []
      coverpage=soup.find_all('loc')
      for link in coverpage:
        if link.get_text().find('mundo')>=0 :
          rows.append(link.get_text()) 
        else :
          if link.get_text().find('union-europea')>=0:
            rows.append(link.get_text()) 
          else :
            if link.get_text().find('politica')>=0:
              rows.append(link.get_text()) 
    except:
      rows = 'error'
      
    return rows
```

```python
extract_urls_topic('https://www.noticiasdegipuzkoa.eus/sitemap_contents_2022_08_227e3_001.xml')
```

```python
pandarallel.initialize(nb_workers= 30, progress_bar=True)

try:
  coverpage = requests.get('https://www.noticiasdegipuzkoa.eus/sitemap_index_227e3.xml').content
  soup = BeautifulSoup(coverpage, 'html.parser')
  
  rows =[]
  coverpage=soup.find_all('loc')
  for link in coverpage:
    rows.append(link.get_text()) 
except:
  rows = 'error'
df = pd.DataFrame(range(0,167),columns = ['page'])
df['page']=rows
print(df)

df['rows']=df['page'].parallel_apply(extract_urls_topic)
df = df[df.rows.str.len()!=0].reset_index(drop=True)
df = df.explode('rows', ignore_index=True)
df=df.drop_duplicates(subset='rows')
df = df.dropna()
print(df)
df.to_csv('/content/drive/MyDrive/Colab Notebooks/noticiasdegipuzkoa/url.csv',index=False)
```
